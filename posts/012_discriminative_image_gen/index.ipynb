# Did OpenAI Push A Discriminatory Model to Prod? 
_Probably, lol_

With all this discussion on AI Image generation, I was inspired by another class I'm taking called CIS 523: Ethical Algorithms to investigate if I could identify instances of discrimination and bias in the new OpenAI image generation model.

## Racial Stereotyping

I went ahead and threw these 3 prompts into GPT and here were the outputs:

### Chinese-American


### Mexican-American


### African-American


**Evidently, discrimination is well and alive in this model.** I don't feel like I need to explain why, just look at the images.

This proves something larger about these LLMs though; their multimodal functionalities espouse the same discrimination, just in different mediums. When crafting the prompt, I tried to make it clear that the person was _**"-American"**_ as well, but the model still made assumptions about their meals. 

Interestingly enough, the image generation for the African-American eating prompt was significantly longer. This is just a hunch, but I feel like there may have been purposeful guardrails put in place relating prompts that involve African-Americans. This could be the case because there is probably more training data relating to those populations, which require some more finesse, particularly in the cultural context of America and big tech (see: [Google Faulted for Racial Biase in Image Search Results for Black Teenagers](https://www.washingtonpost.com/news/morning-mix/wp/2016/06/10/google-faulted-for-racial-bias-in-image-search-results-for-black-teenagers/))
