# Background
As we've discussed in class, a LLM is a giant prediction model, making deep learned connections between large swathes of data. In the particular case of text, there is a lot of text out there on the internet that's discriminatory, and when data like that finds its way into the training set... things like the above happen.

# So How Do We Fix This?
First and foremost, better data. It's important to feed quality data into models to produce good results. The principle of "Garbage in, Garbage out" stands true. Other solutions can be guardrail models that protect against certain prompts and outputs from happening. However, even with precautions, state of the art models like ChatGPT still fall short.
